# Bivariate Choropleth Maps

Hi there!

(image of a bivariate map)

These maps are essentially of two uni-variate choropleth maps. Of course, we can always just map the following these two variables individually,


(image of two choropleths side by side)

But bivariate maps are often super useful for highlighting areas of correlation (or lack thereof) between two variables. This is often an important an important expoloritory analysis step before more sophisticated spatial modelling.

Suprisingly (to me at least), that despite centuries of maps showing multi-variate data, bivariate choropleths like these have only gained popularity in the 1970s via maps created by the U.S. Bureau of the Census. Here's one of those maps. I believe these were the first published bivariate maps (but if you are reading this, and know of earlier bivariate maps, let me know! I'd be super interested to see it).

 In this tutorial, we're going to learn how to create bivariate choropleth maps in [Python](https://www.python.org/) and [QGIS](https://www.qgis.org/en/site/), with some final touch-ups and legend-design in [Inkscape](https://inkscape.org/).
